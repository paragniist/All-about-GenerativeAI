```python
from openai import OpenAI

client = OpenAI(api_key="YOUR_KEY")

response = client.chat.completions.create(
 model="gpt-3.5-turbo",
 messages=[
  {"role":"system","content":"You are helpful"},
  ## System messages are used to set the behavior of the assistant. In this case, we are telling the model to be helpful.
  ## "content" keyword is used to send the message to the model for text generation.
  {"role":"user","content":"Hello"}
  ## user messages are used to send the user's input to the model. In this case, we are sending "Hello" as the user's message.
 ]
)

print(response.choices[0].message.content)
## The response from the model is stored in the "response" variable. We can access the generated message using "response.choices[0].message.content". In this case, we are printing the generated message to the console.
```

```python
from openai import OpenAI

client = OpenAI(
api_key= "Api key",
base_url = "https://generativelanguage.googleapis.com/v1beta/openai/")

response = client.chat.completions.create(
    model="gemini-2.5-flash",
    messages=[
        {"role": "system", "content": "You are a helpful assistant."},
        {"role": "user", "content": "Explain how AI works in simple terms."
        }
    ]
)
print(response.choices[0].message.content)
```

```python
## Open AI format

import json
import requests

API_KEY = "OpenAi key here"
URL = "openai api url here"

headers = {
    "Authorization": f"Bearer {API_KEY}",
    "Content-Type": "application/json"
}
payload = {
    "model ": "gemini-2.5-flash",
    "messages": [
        {
            "role": "system","content": "You are a helpful assistant."
        },
        {
            "role": "user","content": "Explain how AI works in simple terms."
        }
    ]
}   

response = requests.post(URL, headers=headers, data=json.dumps(payload))

#parse the response and print the generated content
data = response.json()

# print the generated content
print(data["choices"][0]["message"]["content"])

## The above code is an example of how to use the OpenAI API to generate text using a chat completion model. We first import the necessary libraries and set up the API key and URL. We then define the headers and payload for the API request, which includes the model we want to use and the messages we want to send to the model. Finally, we make the POST request to the API and parse the response to print the generated content.
```

this code is manual OpenAI-style API calling using requests. No SDK. Raw HTTP.
