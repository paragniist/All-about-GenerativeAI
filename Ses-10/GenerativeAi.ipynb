## Generative Ai:

generative ai refers to the class of artificial intelligence systems designed to create new content such as 
* Text
* images
* Audio
* Video
* Code

based on patterns learned from existing data. 


### Applications: 
* Text generation (eg Gpt, Gemini for conversation)
* Images and art creation (eg DALL-E,Midjourney)
* Music and Video synthesis
* Code Generation(Eg Github Copilot)
