```python
import os
from langchain_openai import ChatOpenAI
from langchain.chains import ConversationChain
from langchain.memory import ConversationBufferMemory
import gradio as gr
```

```python
os.environ['OPENAI_API_KEY'] = ''

llm = ChatOpenAI()

memory = ConversationBufferMemory() # It stores every message in the conversation (both from the human and the AI) in a list.

#This creates a Chain, which is a wrapper that connects the LLM and the memory together into a single functional unit.
conversation = ConversationChain(
    llm=llm,
    memory=memory
)
```

```python
with gr.Blocks() as chat_system:
    chat = gr.Chatbot()
    prompt = gr.Textbox(placeholder="What's on your mind?")
    clear = gr.ClearButton([prompt, chat])
    clear.click(conversation.memory.clear)

    def llm_reply(prompt, chat_history):
        reply = conversation.predict(input=prompt)
        chat_history.append((prompt, reply))
        return "", chat_history

    prompt.submit(llm_reply, [prompt, chat], [prompt, chat])

chat_system.launch()
```
